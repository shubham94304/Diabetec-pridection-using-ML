```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset = pd.read_csv('/content/diabetes (1).csv')
print (diabetes_dataset)
```

```python
x = diabetes_dataset.drop(columns= 'Outcome', axis = 1)
y = diabetes_dataset['Outcome']
```

```python
print(x)
print(y)
```

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_diabetes_dataset = scaler.transform(x)
print(standardized_diabetes_dataset)
```

```python
X = standardized_diabetes_dataset
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

```python
[X_train, X_test, Y_train, Y_test] = train_test_split(X,Y,train_size=0.8, stratify = Y, random_state =2)
print(X.shape, X_test.shape)
```

```python
#LOAD/IMPORTANT MODEL
classifier = svm.SVC(kernel = 'linear')#support vector classfier
```

```python
# NOW ,TRAIN MODEL ON TRAINNIG DATA
classifier.fit(X_train, Y_train)
```

```python
# accuracy score on training data
x_train_data_predict = classifier.predict(X_train)
x_train_data_accuracy = accuracy_score(x_train_data_predict, Y_train)
print('SVM Model accuracy on training data is :',x_train_data_accuracy)
```

```python
# accuracy score on training data
x_test_data_predict = classifier.predict(X_test)
x_test_data_accuracy = accuracy_score(x_test_data_predict, Y_test)
print('SVM Model accuracy on training data is :',x_test_data_accuracy)
```

```python
unseen_data = (10,168,74,0,0,38,0.537,34)
unseen_data_as_numpy_array = np.asarray(unseen_data)
unseen_data_reshape = unseen_data_as_numpy_array.reshape(1,-1)
std_unseen_data = scaler.transform(unseen_data_reshape)
predict = classifier.predict(std_unseen_data)
print(predict)
if(predict[0] == 0):
  print('The person is non-diabetic')
else :
  print('The person is diabetic')
```
